```python
print('1)덧셈 2)뺄셈 3)곱셈 4)나눗셈')
menu = int(input('어떤 연산을 원하는지 번호를 입력하시오:'))
a, b = map(int, input('연산을 원하는 숫자 두개를 입력하시오').split())
sol = 0
if menu == 1:
  menu = '+'
  sol = a + b
elif menu == 2:
  menu = '-'
  sol = a - b
elif menu == 3:
  menu = '*'
  sol = a * b
elif menu == 4:
  menu = '/'
  sol = a / b
print(a, menu, b, '=', sol)
```
